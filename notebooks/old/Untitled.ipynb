```python
import os
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

def analyze_infection_rate(settings):
    
    from spacr.utils import annotate_conditions, save_settings, _map_wells
    from spacr.io import _read_db
    from spacr.settings import set_analyze_class_proportion_defaults
    from spacr.plot import plot_plates, plot_proportion_stacked_bars
    from spacr.stats import perform_normality_tests, perform_levene_test, perform_statistical_tests, perform_posthoc_tests
    from spacr.plot import spacrGraph
    
    def split_filename_column(df, file_column):
        # Extract components using regular expressions
        df['plate'] = df[file_column].str.extract(r'^(plate\d+)')
        col_letter = df[file_column].str.extract(r'_(\w)(\d{2})')[0]  # Extract column letter
        row_number = df[file_column].str.extract(r'_(\w)(\d{2})')[1]  # Extract row number
        df['field'] = df[file_column].str.extract(r'_(\d+)\.npy$')[0]  # Extract field
        # Convert column letters to c1, c2, etc.
        df['column_name'] = 'c' + col_letter.apply(lambda x: str(ord(x.upper()) - ord('A') + 1))
        # Convert row numbers to r1, r2, etc.
        df['row_name'] = 'r' + row_number.astype(int).astype(str)
        return df
    
    #settings = set_analyze_class_proportion_defaults(settings)
    #save_settings(settings, name='analyze_class_proportion', show=True)
    output = {}

    # Process data
    if not isinstance(settings['src'], list):
        settings['src'] = [settings['src']]
    
    locs = []
    for s in settings['src']:
        loc = os.path.join(s, 'measurements/measurements.db')
        locs.append(loc)
            
    dfs = []
    for idx, loc in enumerate(locs):
        print(loc)
        [db_dfs] = _read_db(db_loc=loc, tables=['pivoted_counts'])
        if settings['change_plate']:
            db_dfs['plate'] = f'plate{idx+1}'
            db_dfs['prc'] = db_dfs['plate'].astype(str) + '_' + db_dfs['row_name'].astype(str) + '_' + db_dfs['column_name'].astype(str)
        dfs.append(db_dfs)
            
    df = pd.concat(dfs, axis=0)
    split_filename_column(df, file_column='file_name')
    df['prc'] = df['plate'].astype(str) + '_' + df['row_name'].astype(str) + '_' + df['column_name'].astype(str)
    
    df = annotate_conditions(
        df=df, 
        cells=settings['cell_types'], 
        cell_loc=settings['cell_plate_metadata'], 
        pathogens=settings['pathogen_types'],
        pathogen_loc=settings['pathogen_plate_metadata'],
        treatments=settings['treatments'], 
        treatment_loc=settings['treatment_plate_metadata']
    )
    
    if settings['group_column'] not in df.columns:
        print(f"{settings['group_column']} not found in DataFrame, please choose from:")
        for col in df.columns:
            print(col)
    
    display(df)
    df = df.dropna(subset=[settings['group_column']])
    df['infection_rate'] = df['cell_before_filtration']/df['pathogen_before_filtration']
    
    dst = os.path.join(settings['src'][0], 'results', 'infection_rate')
    os.makedirs(dst,exist_ok=True)
    spacr_graph = spacrGraph(
        df=df,                                       
        grouping_column=settings['group_column'],
        data_column='infection_rate',        
        graph_type=settings['graph_type'],          
        graph_name='test',          
        summary_func='mean',                        
        colors=None,                                 
        output_dir=dst,                              
        save=settings['save'],                       
        y_lim=None,                     
        log_y=False,                    
        log_x=False,                  
        error_bar_type='std',                        
        representation=settings['level'],
        theme='muted',                     
    )
    

    # Create the plot
    spacr_graph.create_plot()

    # Get the figure object if needed
    fig = spacr_graph.get_figure()
    plt.show()

    
    

settings = {
    'src': '/nas_mnt/carruthers/Einar/tsg101_screen/TSG101SCREEN_20240826_140251/plate4',
    'cell_types': ['Hela'],
    'cell_plate_metadata': None,
    'pathogen_types': ['nc', 'pc'],
    'pathogen_plate_metadata': [['c1'], ['c2']],
    'treatments': None,
    'treatment_plate_metadata': None,
    'min_area_bin': 500,
    'group_column': 'pathogen',
    'graph_type':'jitter_bar',
    'pathogen_limit': 1,
    'nuclei_limit': 1000,
    'level': 'well',
    'um_per_px':0.1,
    'change_plate':False,
    'max_bins':5,
    'save':True,
    'verbose': True}

analyze_infection_rate(settings)    
```
